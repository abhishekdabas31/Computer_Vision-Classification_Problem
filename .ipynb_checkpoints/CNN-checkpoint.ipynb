# Convolutional Neural Network

CNN have blockes of Convolutions + Pooling

- A **filter** is multiplied with the input image and it results in a single output. The input has similar number of channels and fewer number of rows than input image. This filter is repreatedly applied to the input image resulting in 2-D output map of activations called **feature maps**.
> lets see it with an example:

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
import skimage

from skimage import color
from scipy import ndimage

%matplotlib inline
```

```python
orig = skimage.color.rgb2gray(skimage.data.moon())
```

```python
#  filters
L_x = [[1, 0, -1],
       [2, 0, -2],
       [1, 0, -1]]
L_y = [[ 1,  2,  1],
       [ 0,  0,  0],
       [-1, -2, -1]]
```

```python
edge_lr = ndimage.convolve(orig, L_x)
edge_ud = ndimage.convolve(orig, L_y)
```

```python
def show_images(images):
  plt.figure(figsize=(10, 10))
  for i, image in enumerate(images):
    plt.subplot(1, len(images), i+1)
    plt.imshow(image, cmap=plt.cm.gray)
    plt.xticks([])
    plt.yticks([])
  plt.tight_layout()
  _ = plt.show()

show_images([orig, edge_lr, edge_ud])
```

## Using Keras for the training the CNN model

```python
import datetime
import os

import tensorflow as tf
import tensorflow.keras as keras
```

```python
(Xtrain, ytrain), (Xtest, ytest) = keras.datasets.mnist.load_data()
```

```python
Xtrain = Xtrain.reshape((-1, 28, 28, 1)).astype(np.float32)
Xtest = Xtest.reshape((-1, 28, 28, 1)).astype(np.float32)
```

```python
# Normalizing the data
Xtrain /= 255
Xtest /= 255
```

```python
ytrain = keras.utils.to_categorical(ytrain, 10)
ytest = keras.utils.to_categorical(ytest, 10)
```

```python
Xtrain.shape, ytrain.shape, Xtest.shape, ytest.shape
```

A CNN is composed of multiple layers of convolution and pooling. A typical convolution block consists of the following.

- Convolutions -- sort of like spatial attention, extract and summarize features from a group of spatially local pixels.
- Nonlinearity -- some kind of activation (usually ReLU).
- Pooling -- downsampling the features extracted from the convolutions across larger and larger blocks of the input.

```python
def build_convpool_unit(layer_id, num_filters, filter_size, pool_size, input_shape=None):
    if input_shape is not None:
    conv = keras.layers.Conv2D(num_filters, filter_size,
                               input_shape=input_shape)
    else:
    conv = keras.layers.Conv2D(num_filters, filter_size)
    return keras.models.Sequential([
    conv,
    keras.layers.Activation("relu"),
    keras.layers.MaxPool2D(pool_size=pool_size)
    ], name="conv_pool_{:d}".format(layer_id))

model = keras.models.Sequential()
model.add(build_convpool_unit(0,20,(5,5),(2,2), input_shape=(28,28,1)))
model.add(build_convpool_unit(1,50,(5,5),(2,2)))
model.add(keras.layers.Flatten(name="flatten"))
model.add(keras.layers.Dense(500, activation = "relu", name="dense_0"))
model.add(keras.layers.Dense(10,activation="softmax",name="dense_1"))
```

```python
model.compile(loss="categorical_crossentropy",
             optimizer="adam",
             metrics=["accuracy"])

model.summary()
```

```python
history = model.fit(Xtrain, ytrain,
                    batch_size=128, 
                    epochs=5, 
                    validation_split=0.2)
```

```python
score = model.evaluate(Xtest, ytest)
print("score: {:.5f}".format(score[0]))
print("accuracy: {:.5f}".format(score[1])) 
```
